# Mapping migration

Introduction to vector data operations

The Veery thrush (Catharus fuscescens) migrates each year between
nesting sites in the U.S. and Canada, and South America. Veeries are
small birds, and encountering a hurricane during their migration would
be disasterous. However, Ornithologist Christopher Hechscher has found
in tracking their migration that their breeding patterns and migration
timing can predicting the severity of hurricane season as accurately as
meterological models (Heckscher 2018)!

1. Some reason animals migrate are to find food, optimal breeding areas, or warmer/colder climates.
2. Climate change may affect animal migration patterns by causing earlier or later departures for suitable temperatures; changing the best places to find food or safe breeding areas because of ecosystem degradation; or making routes longer or less optimal because of more varied and extreme weather events. 
3. I know that deer and elk migrate annually through northern colorado and mountain lions follow them for food. Many species of birds migrate out of Boulder when food becomes scarce in fall/winter, and come back to raise chicks in the spring/summer.
